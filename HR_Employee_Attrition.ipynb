# EDA

## Univariate Analysis

```python
cat_col = []

for i in HR:
    if len(HR[i].unique())<=20:
        cat_col.append(i)
```

```python
cat_col
```

```python
len(cat_col)
```

```python
cont_col = []
for i in HR:
    if len(HR[i].unique()) > 20:
        cont_col.append(i)
```

```python
cont_col
```

```python
len(cont_col)
```

```python
HR_cat = HR[cat_col]
HR_cat.head()
```

```python
HR_cont = HR[cont_col]
HR_cont.head()
```

```python
plt.figure(figsize = (10,30))
plotnumber = 1
for i in HR_cat:
    plt.subplot(13,2, plotnumber)
    sns.countplot(x = HR_cat[i])
    plt.xticks(rotation = 20)
    plotnumber = plotnumber + 1
plt.tight_layout()
```

```python
plt.figure(figsize = (10,30))
plotnumber = 1
for i in HR_cont:
    plt.subplot(8, 3, plotnumber)
    sns.histplot(x = HR_cont[i])
    plotnumber = plotnumber + 1
plt.tight_layout()
```

# Insights from univariant analysis
#### *70% of the people travel raraely, 20% travel frequently rest do not travel.

#### *More than 60% employees seem to be satisfied with their job.
#### *50% of the people are married, 30% single and the rest are divorced.

#### *40% of the people have worked for less than 1 company which implies they are freshers.

# Bivariate Analysis

```python
plt.figure(figsize = (10,30))
plotnumber = 1
for i in HR_cont:
    plt.subplot(8, 3, plotnumber)
    sns.histplot(x = HR_cont[i], hue = HR.Attrition)
    plotnumber = plotnumber + 1
plt.tight_layout()
```
